```python
from geopy.geocoders import Nominatim

def get_coordinates(place_name):
    known_places ={
        "jayanagar": [12.9250, 77.5938],
        "mg road": [12.9716, 77.5946],
        "whitefield": [12.9698, 77.7500],
        "indiranagar": [12.9784, 77.6408],
        "electronic city": [12.8452, 77.6602],
        "majestic": [12.9719, 77.5770],
        "koramangala": [12.9279, 77.6271],
        "hebbal": [13.0354, 77.5988]
    }

    clean_name = place_name.lower().strip()

    if clean_name in known_places:
        return known_places[clean_name]
    try:
        geolocator = Nominatim(user_agent="path_rakshak");
        location = geolocator.geocode(f"{place_name}, Bangalore");
        if location:
            return [location.latitude, location.longitude];
    except:
        pass
    return None #couldnt find 
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Filter out rows where location is 'Other'
mask = ~df2['location'].str.contains('Other')
y_true = df2.loc[mask, 'location']
y_pred = df2.loc[mask, 'estimatedWardName']

# Metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='macro', zero_division=0)
recall = recall_score(y_true, y_pred, average='macro', zero_division=0)
f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)

print(f"Accuracy: {accuracy*100:.2f}%")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")

# Confusion Matrix, prettified
labels = np.unique(y_true)
cm = confusion_matrix(y_true, y_pred, labels=labels)
cm_df = pd.DataFrame(cm, index=labels, columns=labels)

plt.figure(figsize=(12,10))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Ward')
plt.ylabel('Actual Ward')
plt.title('Confusion Matrix: Actual vs Predicted Ward Name')
plt.show()
```

```python
kmeans = KMeans(n_clusters=50, random_state = 42)
# This groups all 1000 accidents into 50 clusters
df2['cluster_id'] = kmeans.fit_predict(df2[['latitude','longitude']])


hotspot_db={}

for i in range(100):
    # get all datas for specific cluster
    cluster_data = df2[df2['cluster_id']==i]

    if len(cluster_data) == 0: continue

    # calculate Stats
    center = kmeans.cluster_centers_[i]   # gets the center of each cluster
    total_accidents = len(cluster_data)   # no. of accidents at each cluster/hotspot 
    avg_speed = cluster_data['speed'].mean()   # avg speed at each hotspot
    # print(total_accidents)
    
    # Most Common Risk Type
    primary_risk = cluster_data['Risk_Category'].mode()[0]   # 0    High Speed/Junction  Name: Risk_Category, dtype: object
    # print(len(primary_risk))

    peak_hour = cluster_data['time'].mode()[0]    # peak hour at each hotspot/cluster

    severity = 'low'
    if total_accidents > 1000: severity='Medium'
    if total_accidents > 3000: severity='High'

    # Store in Hotspot Db
    hotspot_db[i] = {
        'location' : center,
        'risk_level': severity,
        'primary_hazard': primary_risk,
        'avg_speed': round(avg_speed,1),
        'peak_hour': peak_hour,
        'accident_count': total_accidents
    }
    
print("Hotspot Model Trained. Database Created with 50 Zones.")
```

```python
def get_trip_advice(user_lat, user_lon, user_time_hour, user_mode):

    # 1. Find Closest hostpot
    distances = []
    for hid, stats in hotspot_db.items():
        h_lat, h_lon = stats['location']
        dist = np.sqrt((user_lat-h_lat)**2 + (user_lon-h_lon)**2)
        distances.append((dist, hid))
        
    distances.sort();  # sorte on the basis of dist
    closest_dist, closest_id = distances[0]

    # Retrieve the personality of the hotspot
    hotspot = hotspot_db[closest_id]

    # 2. Now make the checks
    risk_score = 5.0
    is_peak_time = (abs(int(hotspot['peak_hour']) - user_time_hour) <=1 )  # around +- 1 hr

    if hotspot['risk_level'] == 'High': risk_score += 2.0
    if is_peak_time: risk_score += 1.5
    if user_mode == 'Motorcycle' and hotspot['primary_hazard']=='High Speed/Junction': risk_score+=1.5
    if user_mode == 'Walking' and hotspot['primary_hazard']=='Pedestrian Conflict': risk_score+=1.5

    # Capp score at 100%
    risk_score = min(risk_score,10.0)

    
    print("\n" + "="*40)
    print(f"TRIP ANALYSIS DASHBOARD")
    print("="*40)
    
    print("\n--- CATEGORICAL SUGGESTIONS (What & Why) ---")
    print(f"1. Overall Trip Risk:  {'HIGH' if risk_score > 7 else 'MODERATE'}")
    print(f"2. Route Summary:      Passes through Hotspot Zone #{closest_id}")
    print(f"3. Primary Hazard:     {hotspot['primary_hazard']}")
    
    if user_mode == 'Motorcycle' and 'Speed' in hotspot['primary_hazard']:
        print(f"4. Mode Warning:       ALERT: High risk for Motorcycles in this zone.")
    else:
        print(f"4. Mode Warning:       Standard caution for {user_mode}.")
        
    if is_peak_time:
        print(f"5. Time Warning:       You are traveling during Peak Risk Hours (approx {hotspot['peak_hour']}:00).")
    else:
        print(f"5. Time Warning:       Time is favorable (Peak risk is {hotspot['peak_hour']}:00).")
        
    print(f"6. Recommendation:     {'Alternative Route Recommended' if risk_score > 8 else 'Proceed with Caution'}")
    print(f"7. Pre-emptive:        {'Defensive Driving Required' if risk_score > 7 else 'Stay in Lane'}")
    print(f"8. Equipment:          {'High-Visibility Gear Required' if user_time_hour > 18 else 'Standard Gear'}")

    print("\n--- NUMERICAL SUGGESTIONS (How Much) ---")
    print(f"1. Overall Risk Score: {risk_score} / 10")
    print(f"2. Hotspot Details:    Zone #{closest_id} had {hotspot['accident_count']} accidents.")
    print(f"3. Avg Incident Speed: {hotspot['avg_speed']} km/h")
    print(f"4. Rec. Speed Limit:   {max(20, hotspot['avg_speed'] - 10)} km/h")
    print(f"5. Risk Multiplier:    {round(risk_score/3.0, 1)}x average")
    print("="*40 + "\n")

# --- TEST THE MODEL ---
# Inputs: Lat, Lon, Time (24h format), Vehicle Mode
user_start = "Jayanagar"
user_dest = "JP Nagar"
user_time = 18
user_mode = 'Motorcycle'

# Concverting to LATITUDE and LONGITUDES
start_coordinates = get_coordinates(user_start.lower().strip())
dest_coordinates = get_coordinates(user_dest.lower().strip())

if start_coordinates == None or dest_coordinates == None:
    print("Error: We couldn't find that location. Please try a major landmark.")
else:
    # Running the risk model
    # using dest[0] - lats , dest[0] - long
    print(f"Analyzing route to {user_dest} @ {dest_coordinates}...")
    get_trip_advice(dest_coordinates[0], dest_coordinates[1], user_time, user_mode)


# get_trip_advice(12.95, 77.60, 18, "Motorcycle")
```

```python
kmeans = KMeans(n_clusters=50, random_state = 42)
# This groups all 1000 accidents into 50 clusters
df['cluster_id'] = kmeans.fit_predict(df[['latitude','longitude']])


hotspot_db={}

for i in range(50):
    # get all datas for specific cluster
    cluster_data = df[df['cluster_id']==i]

    if len(cluster_data) == 0: continue

    # calculate Stats
    center = kmeans.cluster_centers_[i]   # gets the center of each cluster
    total_accidents = len(cluster_data)   # no. of accidents at each cluster/hotspot 
    avg_speed = cluster_data['speed'].mean()   # avg speed at each hotspot
    
    # Most Common Risk Type
    primary_risk = cluster_data['Risk_Category'].mode()[0]   # 0    High Speed/Junction  Name: Risk_Category, dtype: object
    # print(len(primary_risk))

    peak_hour = cluster_data['hour'].mode()[0]    # peak hour at each hotspot/cluster

    severity = 'low'
    if total_accidents > 20: severity='Medium'
    if total_accidents > 40: severity='High'

    # Store in Hotspot Db
    hotspot_db[i] = {
        'location' : center,
        'risk_level': severity,
        'primary_hazard': primary_risk,
        'avg_speed': round(avg_speed,1),
        'peak_hour': peak_hour,
        'accident_count': total_accidents
    }
    
print("Hotspot Model Trained. Database Created with 50 Zones.")
```

```python
def get_trip_advice(user_lat, user_lon, user_time_hour, user_mode):

    # 1. Find Closest hostpot
    distances = []
    for hid, stats in hotspot_db.items():
        h_lat, h_lon = stats['location']
        dist = np.sqrt((user_lat-h_lat)**2 + (user_lon-h_lon)**2)
        distances.append((dist, hid))
        
    distances.sort();  # sorte on the basis of dist
    closest_dist, closest_id = distances[0]

    # Retrieve the personality of the hotspot
    hotspot = hotspot_db[closest_id]

    # 2. Now make the checks
    risk_score = 5.0
    is_peak_time = (abs(hotspot['peak_hour'] - user_time_hour) <=1 )  # around +- 1 hr

    if hotspot['risk_level'] == 'High': risk_score += 2.0
    if is_peak_time: risk_score += 1.5
    if user_mode == 'Motorcycle' and hotspot['primary_hazard']=='High Speed/Junction': risk_score+=1.5
    if user_mode == 'Walking' and hotspot['primary_hazard']=='Pedestrian Conflict': risk_score+=1.5

    # Capp score at 100%
    risk_score = min(risk_score,10.0)

    
    print("\n" + "="*40)
    print(f"TRIP ANALYSIS DASHBOARD")
    print("="*40)
    
    print("\n--- CATEGORICAL SUGGESTIONS (What & Why) ---")
    print(f"1. Overall Trip Risk:  {'HIGH' if risk_score > 7 else 'MODERATE'}")
    print(f"2. Route Summary:      Passes through Hotspot Zone #{closest_id}")
    print(f"3. Primary Hazard:     {hotspot['primary_hazard']}")
    
    if user_mode == 'Motorcycle' and 'Speed' in hotspot['primary_hazard']:
        print(f"4. Mode Warning:       ALERT: High risk for Motorcycles in this zone.")
    else:
        print(f"4. Mode Warning:       Standard caution for {user_mode}.")
        
    if is_peak_time:
        print(f"5. Time Warning:       You are traveling during Peak Risk Hours (approx {hotspot['peak_hour']}:00).")
    else:
        print(f"5. Time Warning:       Time is favorable (Peak risk is {hotspot['peak_hour']}:00).")
        
    print(f"6. Recommendation:     {'Alternative Route Recommended' if risk_score > 8 else 'Proceed with Caution'}")
    print(f"7. Pre-emptive:        {'Defensive Driving Required' if risk_score > 7 else 'Stay in Lane'}")
    print(f"8. Equipment:          {'High-Visibility Gear Required' if user_time_hour > 18 else 'Standard Gear'}")

    print("\n--- NUMERICAL SUGGESTIONS (How Much) ---")
    print(f"1. Overall Risk Score: {risk_score} / 10")
    print(f"2. Hotspot Details:    Zone #{closest_id} had {hotspot['accident_count']} accidents.")
    print(f"3. Avg Incident Speed: {hotspot['avg_speed']} km/h")
    print(f"4. Rec. Speed Limit:   {max(20, hotspot['avg_speed'] - 10)} km/h")
    print(f"5. Risk Multiplier:    {round(risk_score/3.0, 1)}x average")
    print("="*40 + "\n")

# --- TEST THE MODEL ---
# Inputs: Lat, Lon, Time (24h format), Vehicle Mode
user_start = "Jayanagar"
user_dest = "MG Road"
user_time = 18
user_mode = 'Motorcycle'

# Concverting to LATITUDE and LONGITUDES
start_coordinates = get_coordinates(user_start.lower().strip())
dest_coordinates = get_coordinates(user_dest.lower().strip())

if start_coordinates == None or dest_coordinates == None:
    print("Error: We couldn't find that location. Please try a major landmark.")
else:
    # Running the risk model
    # using dest[0] - lats , dest[0] - long
    print(f"Analyzing route to {user_dest} @ {dest_coordinates}...")
    get_trip_advice(dest_coordinates[0], dest_coordinates[1], user_time, user_mode)


# get_trip_advice(12.95, 77.60, 18, "Motorcycle")
```
